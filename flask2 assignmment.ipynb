### Q2. Why is request used in Flask?

```python
#The request object in Flask is used to handle incoming HTTP requests from clients. It provides access to the data sent by the client, such as form data, query parameters, request headers, and more. The request object allows Flask applications to interact with and process the data sent by the client.
```

### Q3. Why is redirect() used in Flask?

```python
#The redirect() function in Flask is used to redirect the user's browser to a different URL. It instructs the client to make a new request to the specified URL, causing the browser to navigate to that URL.
```
